```python
# ------ Main Class ------
def Scheduling(Cwl, Tm,E):
    for i in range(len(Cwl)):
        rdTeam = random.randint(0,len(Tm))
        rdSlot = random.randint(0,len(Tm[0])-1)
        rdEmp = random.randint(0,len(E.Ename))
        while (Tm[rdTeam, rdSlot] != "xxx"):
            rdTeam = random.randint(0,len(Tm)-1)
            rdSlot = random.randint(0,len(Tm[0])-1)
            rdEmp = random.randint(0,len(E.Ename))
        Cwl[i,1] = rdTeam
        Cwl[i,2] = rdSlot
        Cwl[i,5] = rdEmp
        Cwl[i,6] = rdEmp
        if rdSlot == 1 or rdSlot == 2:
            Cwl[i,4] = "จันทร์"
        elif rdSlot == 3 or rdSlot == 4:
            Cwl[i,4] = "อังคาร"
        elif rdSlot == 5 or rdSlot == 6:
            Cwl[i,4] = "พุธ"
        elif rdSlot == 7 or rdSlot == 8:
            Cwl[i,4] = "พฤหัสบดี"
        else:
            Cwl[i,4] = "ศุกร์"
        Tm[int(Cwl[i,1]), int(Cwl[i,2])] = Cwl[i,0]

def ComputeFittnessValue(Cwl) :
    value = 100
    #นับจำนวนงานของแต่ละทีม
    all = []
    for i in range(0,len(Cwl)):
        all.append(Cwl[i][1])
    unique, counts = np.unique(all, return_counts=True)
    
    #หาค่าเฉลี่ยของงาน
    sum = 0
    for i in range(0,len(counts)):
        sum += counts[i]
    sum = sum/len(counts)
    up = int(np.ceil(sum))
    down = int(np.floor(sum))
    
    #เช็คว่าแต่ละทีมมีงานที่เท่าเทียมกัน
    check = 0
    for i in range(0,len(counts)):
        if counts[i] == up or counts[i] == down:
            check += 1 
        else:
            value = 0
    if check == 3:
        value = 100
    return value,counts

def DisplayTable(T,F,C):
    #แสดงตารางงาน
    Tm = T.getTeamWeek()
    print("ตารางทำงาน 5 วันต่อสัปดาห์ ของแต่ละทีมช่าง")
    for i in range(len(Tm)):
        print("Team",T.Tname[i][0],"มี",C[i],"งาน")
        for sl in range(len(Tm[i])):
            print(Tm[i,sl], end=" ")
            if sl%2 == 1 :
                print()
        print()
    print("Fistness Value",F)
    
def DisplayCourseWorkLoad(cwl,T,E):
    #แสดงรายละเอียดงาน
    T = T.Tname
    E = E.Ename
    print("รายละเอียดงานมั้งหมดในสัปดาห์")
    for i in range(len(cwl)):
        print(cwl[i][0],T[int(cwl[i][1])][0],cwl[i][3],cwl[i][4],E[int(cwl[i][5])][0],E[int(cwl[i][6])][1])

def DisplayTeam(T):
    #แสดงทีมช่างทั้งหมด
    print("ทีมช่างในบริษัท")
    print(T.Tname)
    print()
    
# np.random.seed(6)
fitness = 0
while (fitness != 100) :
    T = Teams()
    C = CoursesWorkload()
    E = Employers()
    Scheduling(C.getCourseWorkload(), T.getTeamWeek(),E)
    fitness,counts = ComputeFittnessValue(C.getCourseWorkload())
    
    if (fitness == 100) :
        print("ปริมาณงานของบริษัทมีทั้งหมด",len(C.getCourseWorkload()),"งาน")
        DisplayTable(T,fitness,counts)
```
